```python
#22 sa 24 Tak ka log ziyada order kar ta ha online food
plt.figure()
sns.histplot(df['Age'], bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.countplot(x='Gender', data=df)
plt.title('Gender Comparison')
plt.xlabel('Gender')
plt.ylabel('Number of Users')
plt.show()
```

```python
plt.figure()
sns.barplot(x=df['Monthly Income'])
plt.title('Monthly Income Spread')
plt.show()
```

```python
plt.figure()
plt.hist(df['Monthly Income'], bins=10)
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.countplot(x='Marital Status', data=df)
plt.title('Marital Status vs Orders')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Columns Only)')
plt.show()
```

```python
plt.figure()
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df["Output"].value_counts().plot(kind="bar")
plt.title("Online Food Order (Yes / No)")
plt.xlabel("Order")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df["Age"].plot(kind="hist")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
